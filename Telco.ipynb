# Pandas/Seaborn Data Cleaning & Visualization Project

```python
#Import CSV
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"C:\Users\salva\OneDrive\Escritorio\Portfolio\Python\Telco.csv")
df
```

```python
# drop non use columns
df = df.drop(columns = ["Zip Code", "Lat Long", "Latitude", "Longitude", "Dependents", "Online Security", "Online Backup", "Device Protection","Tech Support", "Streaming TV", "Paperless Billing", "Payment Method"])
df
```

```python
#Contrac Type
sns.set_theme()
plt.figure(figsize=(8,5))
ax=sns.countplot(data=df, x="Contract", palette="viridis")
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Contract Type")
plt.show()
```

```python
#Internet Type
sns.set_theme()
plt.figure(figsize=(8,5))
ax=sns.countplot(data=df, x="Internet Service", palette="viridis")
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Internet Type")
plt.show()
```

```python
#Genders
sns.set_theme()
plt.figure(figsize=(8,5))
ax=sns.countplot(data=df, x="Gender", palette="viridis")
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Genders")
plt.show()
```

```python
#Phone Service
sns.set_theme()
plt.figure(figsize=(8,5))
ax=sns.countplot(data=df, x="Phone Service", palette="viridis")
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Phone Service")
plt.show()
```

```python
#Multiple Lines
sns.set_theme()
plt.figure(figsize=(8,5))
ax=sns.countplot(data=df, x="Multiple Lines", palette="viridis")
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Multiple Lines")
plt.show()
```

```python
#Satisfaction Score
sns.set_theme()
plt.figure(figsize=(8,5))
ax=sns.violinplot(data=df, x="Satisfaction Score", y="Gender", linewidth=1, palette="viridis")
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Satisfaction Score")
plt.show()
```

```python
#Customer Status
sns.set_theme()
plt.figure(figsize=(12, 5))
df_ordered = df.sort_values(by="Customer Status", ascending=False)
ax = sns.countplot(data=df_ordered, x="Customer Status", linewidth=1, palette="viridis")
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Customer Status")
plt.show()
```

```python
#Churn Reasons
sns.set_theme()
plt.figure(figsize=(8, 5))
pivot_table = df.pivot_table(values="Count", index="Churn Reason", aggfunc="sum")
pivot_table = pivot_table.sort_values(by="Count", ascending=False)
ax = sns.heatmap(data=pivot_table,cmap="crest", annot=True, fmt=".2f", linewidths=0.5,)
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Churn Reasons")
plt.show()
```

```python
#Churn Category
sns.set_theme()
plt.figure(figsize=(8, 5))
pivot_table = df.pivot_table(values="Count", index="Churn Category", aggfunc="sum")
pivot_table = pivot_table.sort_values(by="Count", ascending=False)
ax = sns.heatmap(data=pivot_table,cmap="crest", annot=True, fmt=".2f", linewidths=0.5,)
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Churn Category")
plt.show()
```
