# Seaborn Data Visualization Project

```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#Load Dataset
tips = sns.load_dataset("tips")
tips
```

```python
#Rename columns
tips = tips.rename(columns={'sex': 'gender', 'size': 'people'})
tips
```

```python
#Total Bill By Day
sns.set_theme()
ax=sns.barplot(data=tips, x="day", y="total_bill", estimator=sum, palette="viridis")
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Total Bill By Day")
plt.show()
```

```python
#Average Bill By Day
ax=sns.barplot(data=tips, x="day", y="total_bill", estimator=np.mean, palette="viridis")
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Average Bill By Day")
plt.show()
```

```python
#Total Tips By Day
ax=sns.barplot(data=tips, x="day", y="tip", estimator=sum, palette="viridis")
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Total Tips By Day")
plt.show()
```

```python
#Average Tip By Day
ax=sns.barplot(data=tips, x="day", y="tip", estimator=np.mean, palette="viridis")
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Average Tip By Day")
plt.show()
```

```python
#Tip Amount
ax=sns.histplot(data=tips, x="tip", hue="time", multiple="stack", kde=True, palette="viridis")
ax2=sns.rugplot(data=tips, x="tip", height=0.04)
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Tip Amount By Time")
plt.show()
```

```python
#Who Spend Most
sns.set_theme()
plt.figure(figsize=(10,5))
ax=sns.scatterplot(data=tips, x="total_bill", y="tip", hue="gender", style="time", estimator="sum", palette="mako")
ax2=sns.rugplot(data=tips, x="total_bill", y="tip", height=0.04)
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.title("Time Expenses By Gender")
plt.show()
```

```python
#Profitable Days By Gender & Smoker Category
sns.set_theme()
ax=sns.relplot(data=tips, x="total_bill", y="tip", kind="line", col="day", row="gender", height=3.5, hue="smoker", palette="viridis")
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.show()
```
